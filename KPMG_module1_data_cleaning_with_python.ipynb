```python
df2.head()
```

```python
df2.info()
```

```python
# we remove all null values
df2.dropna(how='any',axis=0, inplace = True)
```

```python
df2.info()
```

```python
# concatenate first and last names
# Note that this isnt a must, it is just my choice
df2['full_name'] = df2['first_name'] + " " + df2['last_name']
```

```python
# in this block, we try to chage the index of the 'full_name' columnn
col_name = 'full_name'
first_col = df2.pop(col_name)
df2.insert(3, col_name, first_col)
```

```python
df2
```

```python
df2['gender'].unique()
```

```python
# to change inconsistencies of 'gender' values
df2['gender'].replace({'F':'Female','Femal':'Female'}, inplace = True)
```

```python
df2['gender'].unique()
```

```python
# to convert date of irth to datetime format
df2['DOB'] = pd.to_datetime(df2['DOB'], format = '%Y-%m-%d')
```

```python
today = datetime.today()
```

```python
# create an age column from existing 'DOB' column
df2['age'] = df2['DOB'].apply(lambda x: today.year - x.year -((today.month, today.day)<(x.month, x.day)))
```

```python
# change the index location of 'age'
df2 = df2.reindex(columns=["customer_id",'first_name','last_name','full_name', 'gender','past_3_years_bike_related_purchases','DOB','age','job_title','job_industry_category','wealth_segment','deceased_indicator','default','owns_car','tenure'])
```

```python
df2['deceased_indicator'].unique()
```

```python
indexname = df2[df2['deceased_indicator']=='Y'].index
```

```python
# drop the deceased customers
df2.drop(indexname, inplace = True)
df2['deceased_indicator'].unique()
```

```python
df2.drop('default', axis =1, inplace = True)
```

# done with df2; 'CustomerDemographic'

```python
df1
```

```python
df1.info()
```

```python
# change the '0.0' and '1.0' to 'False' and 'True' respectively
df1['online_order'].replace({0.0:'False',1.0:'True'}, inplace = True)
```

```python
# drop null values
df1.dropna(how='any', axis = 0, inplace = True)
```

```python
df1.info()
```

```python
# add a profit column
df1['profit'] = df1.apply(lambda x: x['list_price'] - x['standard_cost'], axis=1)
```

```python
df1.head()
```

```python
# add a profit percentage column
# this is also optional
df1['profit%'] = df1.apply(lambda x: (x['profit'] / x['list_price'])*100, axis=1)
```

```python
df1['profit%'] = np.round(df1['profit%'], decimals = 2)
```
